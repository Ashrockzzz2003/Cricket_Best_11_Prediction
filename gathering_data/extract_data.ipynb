```python
import pandas as pd
from requests import get
import re
from bs4 import BeautifulSoup
```

## Extract Player Data

```python
df = pd.read_csv('player_names.csv')
df
```

```python
df['team_name'].value_counts()
```

```python
# Add new column PlayerID
df['player_id'] = None
df
```

### Australia

```python
australian_players = df[df['team_name'] == 'Australia']['player_name'].values
australian_players

for player in australian_players:
    # GET REQUEST
    # https://stats.espncricinfo.com/ci/engine/stats/analysis.html?search=Pat+Cummins;template=analysis

    if player == 'Mitch Marsh':
        df.loc[df['player_name'] == player, 'player_id'] = '272450'
        continue

    request_url = f'https://stats.espncricinfo.com/ci/engine/stats/analysis.html?search={player.replace(" ", "+")};template=analysis'

    response = get(request_url)

    soup = BeautifulSoup(response.text, 'html.parser')

    if response.status_code == 200:
        # find all links with className statsLinks
        links = soup.find_all('a', class_='statsLinks')
        ends_with = ".html?class=2;type=allround"
        starts_with = "/ci/engine/player/"
        
        for link in links:
            if link['href'].endswith(ends_with) and link['href'].startswith(starts_with):
                player_id = link['href'].split('/')[-1].split('.')[0]
                df.loc[df['player_name'] == player, 'player_id'] = player_id

df[df['team_name'] == 'Australia']
```
